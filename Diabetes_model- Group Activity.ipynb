```python
#import libraries and import the csv file
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

filename=pd.read_csv('diabetes_model.csv')
df=pd.DataFrame(filename)

import warnings
warnings.filterwarnings("ignore")
```

# 1.count plot of outcome

```python
plt.style.use('seaborn')
plt.figure(figsize=(4,8))
sns.countplot(df['Outcome'])
plt.title('Count plot of Outcome',fontsize=16)
plt.xlabel('Outcome',fontsize=14)
plt.ylabel('Count',fontsize=14)
plt.tight_layout()
```

# 2.Analyse the effect of age on Outcome using Boxplot

```python
plt.style.use('seaborn')
plt.figure(figsize=(4,8))
sns.boxplot(x='Outcome',y='Age',data=df)
plt.title('Effect of Age on Outcome',fontsize=16)
plt.xlabel('Outcome',fontsize=14)
plt.ylabel('Age',fontsize=14)
```

# 3.Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,8))
sns.stripplot(x=df['Outcome'], y=df['DiabetesPedigreeFunction'])
plt.title('Probability of Diabetes in family',fontsize=16)
plt.xlabel('Outcome',fontsize=14)
plt.ylabel('DiabetesPedigreeFunction',fontsize=14)
plt.tight_layout()
```

```python
#from the plot, it is clear that as the DiabetesPedigreeFunction increases, the chances of getting diabetes also increases.
```

# 4.Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,8))
sns.scatterplot(x='Insulin',y='Glucose',data=df,hue='Outcome')
plt.title('Relationship between Insulin & Glucose',fontsize=14)
```

# 5.Create a heatmap on Correlation of features in the data

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(),annot=True)
plt.xticks(rotation=90)
```
