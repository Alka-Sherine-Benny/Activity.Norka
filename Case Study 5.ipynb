# 3. Check whether all the zones generate the same amount of sales.

Let,

Null hypothesis (H0) :- All the zones generate the same amount of sales.

Alternative hypothesis (Ha) :-  All the zones doesnot generate the same amount of sales.

set significance level, alpha as 0.05

if p value <0.05, we could reject the Null hypothesis and accept the Alternative hypothesis.

if p value >0.05, we could accept the null hypothesis and reject the Alternative hypothesis.

```python
from scipy.stats import f_oneway

F, p = f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
print('F-statistic :',F)
print('p-value :',p)
```

```python
if p<0.05:
    print('We reject the Null hypothesis; ie. all the zones generate the same amount of sales. We conclude that all the zones doesnot generate the same amount of sales.')
else:
    print('We accept the Null hypothesis; ie. all the zones doesnot generate the same amount of sales. We conclude that all the zones generate the same amount of sales.')
```
